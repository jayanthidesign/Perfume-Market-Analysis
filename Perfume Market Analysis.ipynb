# **Visualization**

**Top cities by sales**

```python
sales_city=df.groupby('city')['sold'].sum().sort_values(ascending=False).head(5)
plt.figure()
plt.bar(sales_city.index,sales_city.values)
plt.title("Top Cities by Sales")
plt.xlabel("city")
plt.ylabel("sold")
plt.show()
```

**Category Distribution**

```python
Total_sales= df.groupby('category')['sold'].sum()
plt.figure()
plt.pie(Total_sales.values,labels=Total_sales.index, autopct='%1.1f%%')
plt.title('Category Distribution')
plt.show()
```
